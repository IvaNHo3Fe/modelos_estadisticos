```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame()

for i in range(1,50):
  muestra = np.random.binomial(1000,0.6,10)
  col = f"muestra{i}"
  df[col] = muestra

df.head(5)
```

```python
## Necesitamos las medias dentro de otro dataframe
## De esta manera podremos graficar con Seaborn

df_muestra_medias = pd.DataFrame(df.mean(), columns = ["Muestra de las medias"])
df_muestra_medias
```
